# Preprocessing Text

## Load and explore GPT2 data

```python
df_train = pd.read_csv('data/training_set.csv')
df_test = pd.read_csv('data/testing_set.csv')

display(df_train)
```

```python
report_ID = {}
sentence_dict = {}
for i, (ID, report, img) in enumerate(zip(df_train["Patient ID"], df_train["Findings"], df_train["Image Index"])):
    if ID not in report_ID:
        report, token = clean_report(cut_startseq_endseq(report).strip().lower())
        report_ID[ID] = {"report": report, "token": token, "img": []}
        
        for sent in token:
            if sent not in sentence_dict:
                sentence_dict[sent] = 0
            sentence_dict[sent] += 1

    report_ID[ID]["img"].append(img)
```

```python
id_ = 0
sentence_encoder = {}
sentence_decoder = {}
unclean_sent = []
for sent in sentence_dict:
    if len(sent.split()) == 1:
        print(sentence_dict[sent], sent)
        unclean_sent.append(sent)
        print()
    sentence_encoder[id_] = sent
    sentence_decoder[sent] = id_
    id_ += 1
```

```python
pickle.dump(sentence_decoder, open('data/sentence_encoder.pkl', 'wb'))
pickle.dump(sentence_encoder, open('data/sentence_decoder,pkl', 'wb'))
```

# DataLoader

```python
class XRayDataset(Dataset):
    def __init__(self, data, img_dir, transform):
        self.data = data
        self.img_dir = img_dir
        self.transform = transform

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        img_path = os.path.join(self.img_dir, self.data.iloc[idx, 0])
        image = Image.open(img_path)
        label = self.data.iloc[idx, 2]
        
        image = transform_img(image, self.transform)
        _, label = tokenize_report(label)
        return image, label
```

```python
data = XRayDataset(df_train, 'data/images', transform)
```
