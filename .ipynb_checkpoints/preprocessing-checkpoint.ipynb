# Import Library

```python
import torch
import torchvision.transforms as T
import numpy as np
import pandas as pd
import cv2
import os
from PIL import Image
import matplotlib.pyplot as plt
from sentence_transformers import SentenceTransformer
```

# Preprocessing Text

## Load and explore data

```python
df_train = pd.read_csv('data/training_set.csv')
df_test = pd.read_csv('data/testing_set.csv')

print(df_train.head())
```

## Sentence Embedding

```python
##Load Model (Sbert)
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')

#Our sentences we like to encode
sentences = df_train.Impression

#Sentences are encoded by calling model.encode()
embeddings = model.encode(sentences)
```

```python
print(embeddings.shape)
```

# Preprocessing Image

## Load and explore Images

```python
img_path = 'data/images'
filenames = os.listdir(img_path)
f = filenames[0]

img = cv2.imread(os.path.join(img_path, f))
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(16, 10))
plt.imshow(img, cmap='gray', vmin=0, vmax=255)
plt.axis(False)
plt.show()
```

## Normalize and Transforms Images

```python
transform = T.Compose([
    T.Resize(800),
    T.ToTensor(),
    T.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

t_img = transform(Image.fromarray(img))
```
